```python
m = 0
n = 0
a = 0
b = 0
for x in range (len(MasterDataFrame)):
    if MasterDataFrame["Gender"].iloc[x] == "Male":
        m += MasterDataFrame["annual premium (in Rs)"].iloc[x]
        a += 1
    
    else:
        n += MasterDataFrame["annual premium (in Rs)"].iloc[x]
        b += 1
        
    x += 1
```

```python
print("Male Average is",m/a,"and Female Average is",n/b,"\nThe Data is balanced between the genders")
```

```python
r = 0
s = 0
t = 0
u = 0
v = 0
w = 0
e = 0
f = 0
g = 0
h = 0
i = 0
j = 0
for x in range (len(MasterDataFrame)):
    if ((MasterDataFrame["age"].iloc[x] < 30) and (MasterDataFrame["age"].iloc[x] >= 20)):
        r += MasterDataFrame["annual premium (in Rs)"].iloc[x]
        e += 1
        
    elif ((MasterDataFrame["age"].iloc[x] < 40) and (MasterDataFrame["age"].iloc[x] >= 30)):
        s += MasterDataFrame["annual premium (in Rs)"].iloc[x]
        f += 1
        
    elif ((MasterDataFrame["age"].iloc[x] < 50) and (MasterDataFrame["age"].iloc[x] >= 40)):
        t += MasterDataFrame["annual premium (in Rs)"].iloc[x]
        g += 1
        
    elif ((MasterDataFrame["age"].iloc[x] < 60) and (MasterDataFrame["age"].iloc[x] >= 50)):
        u += MasterDataFrame["annual premium (in Rs)"].iloc[x]
        h += 1
        
    elif ((MasterDataFrame["age"].iloc[x] < 70) and (MasterDataFrame["age"].iloc[x] >= 60)):
        v += MasterDataFrame["annual premium (in Rs)"].iloc[x]
        i += 1
        
    elif ((MasterDataFrame["age"].iloc[x] < 80) and (MasterDataFrame["age"].iloc[x] >= 70)):
        w += MasterDataFrame["annual premium (in Rs)"].iloc[x]
        j += 1
        
    x += 1
```

```python
print("Average annual between 10 to 20:",r/e)
print("Average annual between 30 to 40:",s/f)
print("Average annual between 40 to 50:",t/g)
print("Average annual between 50 to 60:",u/h)
print("Average annual between 60 to 70:",v/i)
print("Average annual between 70 to 80:",w/j)
```

```python
o = 0
p = 0
q = 0
c = 0
d = 0
k = 0
for x in range (len(MasterDataFrame)):
    if MasterDataFrame["vehicle age"].iloc[x] == "< 1 Year":
        o += MasterDataFrame["annual premium (in Rs)"].iloc[x]
        c += 1
        
    elif MasterDataFrame["vehicle age"].iloc[x] == "1-2 Year":
        p += MasterDataFrame["annual premium (in Rs)"].iloc[x]
        d += 1
        
    elif MasterDataFrame["vehicle age"].iloc[x] == "> 2 Years":
        q += MasterDataFrame["annual premium (in Rs)"].iloc[x]
        k += 1
        
    x += 1
```

```python
print("Average annual < 1 Year:",o/c)
print("Average annual < 1-2 Year:",p/d)
print("Average annual > 2 Years",q/k)
```
